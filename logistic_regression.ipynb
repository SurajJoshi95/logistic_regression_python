```python
# importing the dataset
import pandas as pd
dataset=pd.read_csv("D:/NP/dataset.csv")
```

```python
# view dataset
dataset
```

```python
# train_test split
from sklearn.cross_validation import train_test_split

train , test=train_test_split(dataset,test_size=0.2)

# spliting tabel
# train dataset
train_x=train.iloc[:,0:24].values
train_label=train.iloc[:,-1].values
train_x=pd.DataFrame(train_x)
train_label=pd.DataFrame(train_label)

# test dataset
test_x=test.iloc[:,0:24].values
test_label=test.iloc[:,-1].values
test_x=pd.DataFrame(test_x)
test_label=pd.DataFrame(test_label)
```

```python
# converting dataset into integer
# train_x
train_x.convert_objects(convert_numeric=True)
test_x.convert_objects(convert_numeric=True)
```

```python
from sklearn.preprocessing import scale
d=train_x
scale(d)
```

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(d,train_label)
#m.summary()
```

```python
# prdictions ( Training set)
y_pred = lr.predict(d)
lr.score(d, train_label)
#y_pred
```

```python
# predictions ( Test set)
z_pred=lr.predict(test_x)
lr.score(test_x,test_label)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(z_pred, test_label)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(z_pred, test_label))
```
